```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
color_pal = sns.color_palette()
```

```python
### Load the dataset
df=pd.read_csv('AEP_hourly.csv')
df.head()
df=df.set_index('Datetime')
df.index=pd.to_datetime(df.index)
```

```python
df.plot(style='.', figsize=(15, 5), title='AEP Hourly Energy Consumption',color=color_pal[0])
plt.show()
```

```python
# Plot the rolling mean of the energy consumption
df['AEP_MW'].rolling(window=24).mean().plot(figsize=(15, 6), color='red')
plt.title('24-Hour Rolling Mean of Energy Consumption')
plt.xlabel('Date')
plt.ylabel('Energy Consumption (MW)')
plt.grid(True)
plt.show()
```

```python
#Monthly Average Consumption
monthly_avg = df['AEP_MW'].resample('M').mean()
plt.figure(figsize=(15,6))
monthly_avg.plot()
plt.title('Monthly Average Energy Consumption')
plt.xlabel('Month')
plt.ylabel('Average Energy (MW)')
plt.grid(True)
plt.show()
```

```python
#Heatmap of missing values (if any)
print(df.isnull().sum())
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False)
plt.title('Heatmap of Missing Values')
plt.show()
```

```python
# Correlation Analysis
print("\nCorrelation Matrix:")
print(df.corr())

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
